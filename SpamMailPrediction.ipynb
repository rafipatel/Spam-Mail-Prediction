```python
#Loading Libraries
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Loading Dataset 
r_data = pd.read_csv('spamham.csv',engine='python')
"""removing null Values"""
mailData = r_data.where((pd.notnull(r_data)),'')
```

```python
mailData.shape
```

```python
mailData.head()
```

```python
#label spam mail as 0 and not spam mail (ham) as 1
mailData.loc[mailData['v1']=="spam","v1",] = 0
mailData.loc[mailData['v1']=="ham","v1",] = 1
```

```python
#X as text and Y as label (0/1) (spam/ham)
X = mailData.v2
Y = mailData.v1

print(X)
print(Y)
```

```python
#spliting data
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,train_size =0.8,test_size = 0.2)
```

Extracting Features** from *text data *so to use it as input to SVM model using Tfidfvectorizer 

```python
#convert text to lower case (better Practice)

Fextract = TfidfVectorizer(min_df=1 , stop_words='english',lowercase ='True')
X_train_features = Fextract.fit_transform(X_train)
X_test_features = Fextract.transform(X_test)

#convert y train and y test values from string to int
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')


```

Training the model --> SVM (very useful for classification and regression problems)

```python
mymodel = LinearSVC()
mymodel.fit(X_train_features,Y_train)

#Evaluationg Model

mmodel = mymodel.predict(X_train_features)
accuracy_train = accuracy_score(Y_train,mmodel)


print(accuracy_train)

```

```python
#For Testing
#by mistakely revealed the testing data to model by using.fit for test stuffs,, haha LOL


Mmodel = mymodel.predict(X_test_features)
accuracy_test = accuracy_score(Y_test,Mmodel)


print(accuracy_test)
```
